# Gradient Descent Optimizer

```python
#import
import tensorflow as tf
```

```python
#the equation is : 4x+2=0
```

```python
tf.compat.v1.disable_eager_execution()

unknowvalue = tf.Variable(0.0)   # x
a =tf.constant(4.0)
b= tf.constant(2.0)
c= tf.multiply(unknowvalue,a)   #4x


equation = tf.add(c,b)  # 4x+2

zerovalue = tf.constant(0.0)
diff = tf.square(equation- zerovalue) # differnce is : 4x+2  -  0 

```

```python
solving = tf.compat.v1.train.GradientDescentOptimizer(0.01).minimize(diff)
init = tf.compat.v1.global_variables_initializer()


sess=tf.compat.v1.Session()
sess.run(init)
for n in range(1000):
    sess.run(solving)
```

```python
print(sess.run(unknowvalue))
```

```python
#  Another  solution

tf.compat.v1.disable_eager_execution()

unknowvalue = tf.Variable(0.0)   # x
a = tf.constant(4.0)
b = tf.constant(2.0)
c = tf.multiply(unknowvalue, a)   #4x
equation = tf.add(c, b)  # 4x+2
zerovalue = tf.constant(0.0)
diff = tf.square(equation - zerovalue) # difference is : 4x+2  -  0 
solving = tf.compat.v1.train.GradientDescentOptimizer(0.01).minimize(diff)
init = tf.compat.v1.global_variables_initializer()

sess = tf.compat.v1.Session()
sess.run(init)
for n in range(1000):
    sess.run(solving)
    print("Iteration", n, ": x =", sess.run(unknowvalue))
print("Final value of x:", sess.run(unknowvalue))
```

```python
#the equation is : -10x+50=0
```

```python
tf.compat.v1.disable_eager_execution()

unknowvalue = tf.Variable(0.0)   # x
a =tf.constant(-10.0)
b= tf.constant(50.0)
c= tf.multiply(unknowvalue,a)   #10x


equation = tf.add(c,b)  # -10x+50

zerovalue = tf.constant(0.0)
diff = tf.square(equation- zerovalue) # differnce is :-10x+50    0 

```

```python
solving = tf.compat.v1.train.GradientDescentOptimizer(0.0001).minimize(diff)
init = tf.compat.v1.global_variables_initializer()


sess=tf.compat.v1.Session()
sess.run(init)
for n in range(1000):
    sess.run(solving)
```

```python
print(sess.run(unknowvalue))
```
